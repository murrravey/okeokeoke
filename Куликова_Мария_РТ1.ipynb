> –ò–∑–º–µ–Ω—è–µ–º—ã–µ –æ–±—ä–µ–∫—Ç—ã:

##1.2 –ü—Ä–∏–º–µ—Ä

```python
x = 3+5.2*7
y = None
z = 'a', 5, 12.345, (2,'b')
df = [['–ê–Ω—Ç–æ–Ω–æ–≤–∞ –ê–Ω—Ç–æ–Ω–∏–Ω–∞', 34, '–∂'],[ '–ë–æ—Ä–∏—Å–æ–≤ –ë–æ—Ä–∏—Å', 26, '–º']]
A = {1, 'title', 2, 'content'}
print(x,'|', type(x),'\n', y, '|', type(y), '\n',
    z, '|', type(z),'\n', df, '|', type(df), '\n', A, '|', type(A))
```

##2.2 –ü—Ä–∏–º–µ—Ä
–í—ã–≤–µ—Å—Ç–∏ –Ω–∞ —ç–∫—Ä–∞–Ω —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è —Ö –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–π,–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–π –∏–ª–∏ —Ä–∞–≤–Ω–∞ –Ω—É–ª—é.

```python
x = 125
if x < 0:
  print("x –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π")
elif x == 0:
  print("x —Ä–∞–≤–Ω–æ –Ω—É–ª—é")
else:
  print("x –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π")
```

##2.3. –ó–∞–¥–∞–Ω–∏–µ
–ù–∞–ø–∏—à–∏—Ç–µ –∫–æ–¥. –ó–∞–¥–∞–µ—Ç—Å—è —Ö, –Ω–∞–ø–µ—á–∞—Ç–∞—Ç—å –∫–∞–∫–æ–º—É –∏–∑ –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤
–ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç: (-infinity, -5), [-5, 5] –∏–ª–∏ –æ—Ç (5, +infinity)

```python
x = 5
if x < -5:
  print("x (-infinity, -5)")
elif x > 5:
  print ("x (5, +infinity)")
else:
  print ("x [-5, 5]")
```

##4.3.1 –ó–∞–¥–∞–Ω–∏–µ
–ó–∞–¥–∞–π—Ç–µ –º–∞—Å—Å–∏–≤ —Å–ª—É—á–∞–π–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∏–∑ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞(0; 1).
–†–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ —Å—Ä–µ–¥–Ω–∏–µ –∏ –º–µ–¥–∏–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è –º–∞—Å—Å–∏–≤–∞, —Å—Ä–∞–≤–Ω–∏—Ç–µ
—Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã, –∫–∞–∫–∏–µ –≤—ã–≤–æ–¥—ã –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –æ –∑–Ω–∞—á–µ–Ω–∏—è—Ö?
–ü–æ—Å—Ç—Ä–æ–π—Ç–µ —Ç–æ—á–µ—á–Ω—É—é –¥–∏–∞–≥—Ä–∞–º–º—É —Ä–∞—Å—Å–µ–Ω–∏—è –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞.


```python
import numpy as np
import matplotlib.pyplot as plt
size = 10
x = range(0, size)
random =  np.random.random_sample(size = size)
average = np.average(random)
median = np.median(random)
print(random, average, median)
if average == median:
  print(" –°—Ä–µ–¥–Ω–∏–µ –∏ –º–µ–¥–∏–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ä–∞–≤–Ω—ã")
elif average > median:
  print(" –°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏—è –±–æ–ª—å—à–µ –º–µ–¥–∏–∞–Ω–Ω–æ–≥–æ")
else:
  print(" –ú–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –±–æ–ª—å—à–µ —Å—Ä–µ–¥–Ω–µ–≥–æ")
fig, ax = plt.subplots(figsize = (10,6))
ax.scatter(x, random)
plt.show()
```

##4.3.2 –ó–∞–¥–∞–Ω–∏–µ
–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è –°–Ω–∏–º–æ–∫ —ç–∫—Ä–∞–Ω–∞ 2023-04-11 152755.png –°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–∏ (ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π –ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ - –ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π.


```python
import numpy as np
import matplotlib.pyplot as plt
#–û–ø—Ä–µ–¥–µ–ª—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é
def my_function(x):
  numerator = np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x)**2)
  denominator = abs(1 - np.sin(x)**3)
  return numerator / denominator + np.log(abs(2*x))
#–°–æ–∑–¥–∞–µ–º –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–∏
x = np.linspace(1, 10, num = 10)
y = my_function(x)
#–í—ã–¥–µ–ª—è–µ–º —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π –ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞
y_slice = y[:5]
#–°—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫–∏
plt.plot(x, y, label = '–û—Å–Ω–æ–≤–Ω–æ–π –º–∞—Å—Å–∏–≤')
plt.scatter(x[:5], y_slice, c = 'r', label = '–°—Ä–µ–∑ –ø–µ—Ä–≤–æ–π –ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## –ó–∞–¥–∞–Ω–∏–µ 4.3.3
–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è –°–Ω–∏–º–æ–∫ —ç–∫—Ä–∞–Ω–∞ 2023-04-11 161522.png –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ –Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ (0,10) —Å —à–∞–≥–æ–º 1 —Å –∑–∞–ª–∏–≤–∫–æ–π –ø–ª–æ—â–∞–¥–∏ –∏ –Ω–∞–π—Ç–∏ —ç—Ç—É –ø–ª–æ—â–∞–¥—å –ø–æ–¥ –Ω–∏–º. –î–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø–ª–æ—â–∞–¥–∏ –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏–∏ trapz(), –≤—ã—á–∏—Å–ª—è—é—â–µ–π –∏–Ω—Ç–µ–≥—Ä–∞–ª –ø–æ –ø—Ä–∞–≤–∏–ª—É —Ç—Ä–∞–ø–µ—Ü–∏–∏.
–î–ª—è –µ–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–π —Ä–∞–±–æ—Ç—ã –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø–æ–¥–∫–ª—é—á–∏—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏:
from scipy.integrate import simps
from numpy import trapz

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapz
x = np.arange(0, 10, 1)
def F(x):
  return np.abs(np.cos(x*np.exp(np.cos(x) + np.log(x+1))))
plt.plot(x, F(x))
plt.fill_between(x, F(x), color = 'blue', alpha = 0.2)
plt.show()

area = trapz(F(x), x)
print("–ü–ª–æ—â–∞–¥—å –ø–æ–¥ –∫—Ä–∏–≤–æ–π:", area)
```
